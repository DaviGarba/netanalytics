```python
from netanalytics.subnetworks import common_subgraph

```

```python
res = []
for fol in folders:
    print(fol)
    files = [join(path+fol, f) for f in listdir(path+fol) if isfile(join(path+fol, f))]
    to_contain = fol.replace('Networks', '')
    to_contain = to_contain[-3:]
    integrated = [i for i in files_integrated if to_contain in i][0]
    A_integr = pd.read_csv(integrated, index_col=0)
    #print(A)
    nodes_integr = A_integr.index.values
    A_integr = A_integr.values
    print(number_of_edges(A_integr))
    
    As = []
    nodes_lists= []
    n_edges_single = []
    #print(files)
    for f in files:
        A, nodes, _= from_edge_list(f, sep=' ', only_adjacency=False)
        n_edges_single.append(number_of_edges(A))
        As.append(A)
        nodes_lists.append(nodes)
        common, nodes, n_edges = common_subgraph(A, A_integr, nodes,nodes_integr)
        res.append([f, n_edges_single, common, nodes, n_edges])
        print(n_edges)
```

```python
with open("salvataggio.pkl", 'wb') as f:
    pkl.dump(res, f)
```

```python
def _from_name_to_index(s):
    if s.startswith('Adapt'):
        return 0
    if s.startswith('Lass'):
        return 1
    if s.startswith('genenet'):
        return 2
    if s.startswith('wg'):
        return 3
    if s.startswith('aracnea'):
        return 4
    if s.startswith('aracnem'):
        return 5
    if s.startswith('clr'):
        return 6
    if s.startswith('mrnetb'):
        return 8
    if s.startswith('mrnet'):
        return 7
    if s.startswith('c3net'):
        return 9
    if s.startswith('Genie3'):
        return 10
   
```
